```python
import pandas as pd
import glob

csv_files = glob.glob("*.csv")

df = pd.concat([pd.read_csv(f) for f in csv_files], ignore_index=True)

print("Total rows:", len(df))
df.head()
```

```python
def row_to_text(row):
    return f"""
Crime Report:
Crime ID: {row.get('Crime ID', 'Unknown')}
Date: {row.get('Month', 'Unknown')}
Reported by: {row.get('Reported by', 'Unknown')}
Area: {row.get('Falls within', 'Unknown')}
Location: {row.get('Location', 'Unknown')}
LSOA: {row.get('LSOA name', 'Unknown')}
Crime Type: {row.get('Crime type', 'Unknown')}
Outcome: {row.get('Last outcome category', 'Unknown')}
Coordinates: Latitude {row.get('Latitude', 'Unknown')}, Longitude {row.get('Longitude', 'Unknown')}
"""
```

```python
df = df.fillna("Unknown")

df = df.drop_duplicates(subset=["Location", "Crime type", "Month"])
df = df.sample(80000, random_state=42)
```

```python
print("Total rows:", len(df))
df.head()
```

```python
df["text"] = df.apply(row_to_text, axis=1)
```

```python
df[["text"]].to_json("crime_processed_data.jsonl", orient="records", lines=True)
```
